```python
date = "2018-01-01"
Text = f"https://apis.data.go.kr/1390802/AgriWeather/WeatherObsrInfo/GnrlWeather/getWeatherTenMinList?serviceKey=GjGnI6AQaoyzpnVpX4m1uCboxkiC1XZ2RyDjjJH5X5Pu4Y4R%2BFzFcWpl54KewmQ1lQnY9jFa%2FeLy9EsMTm4v8g%3D%3D&Page_No=1&Page_Size=100&date={date}&time=1300&obsr_Spot_Nm=%EC%9D%98%EC%84%B1%EA%B5%B0%20%EC%9D%98%EC%84%B1%EC%9D%8D&obsr_Spot_Code=037339B009"

# 해당 Text URL에 있는 데이터 중, 태그가 date, max_Temp, min_Temp, hum, widdir, wind, rain, sun_Time, sun_Qy, condens_Time, soil_Temp 인 데이터를 읽어들이고, 데이터 프레임으로 저장
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_weather_data(date):
    Text = f"https://apis.data.go.kr/1390802/AgriWeather/WeatherObsrInfo/GnrlWeather/getWeatherTenMinList?serviceKey=GjGnI6AQaoyzpnVpX4m1uCboxkiC1XZ2RyDjjJH5X5Pu4Y4R%2BFzFcWpl54KewmQ1lQnY9jFa%2FeLy9EsMTm4v8g%3D%3D&Page_No=1&Page_Size=100&date={date}&time=1300&obsr_Spot_Nm=%EC%9D%98%EC%84%B1%EA%B5%B0%20%EC%9D%98%EC%84%B1%EC%9D%8D&obsr_Spot_Code=037339B009"
    response = requests.get(Text, timeout=10)  # 예: 10초 동안 응답을 기다림
    soup = BeautifulSoup(response.text, 'html.parser')
    date = [item.text for item in soup.find_all('date')]
    max_Temp = [item.text for item in soup.find_all('max_temp')]
    min_Temp = [item.text for item in soup.find_all('min_temp')]
    hum = [item.text for item in soup.find_all('hum')]
    widdir = [item.text for item in soup.find_all('widdir')]
    wind = [item.text for item in soup.find_all('wind')]
    rain = [item.text for item in soup.find_all('rain')]
    sun_Time = [item.text for item in soup.find_all('sun_time')]
    sun_Qy = [item.text for item in soup.find_all('sun_qy')]
    condens_Time = [item.text for item in soup.find_all('condens_time')]
    soil_Temp = [item.text for item in soup.find_all('soil_temp')]
    data = {'date': date, 'max_Temp': max_Temp, 'min_Temp': min_Temp, 'hum': hum, 'widdir': widdir, 'wind': wind, 'rain': rain, 'sun_Time': sun_Time, 'sun_Qy': sun_Qy, 'condens_Time': condens_Time, 'soil_Temp': soil_Temp}
    df = pd.DataFrame(data)
    return df
```

```python
date = "2017-01-01"
# date 변수를 2010-01-01에서 2024-05-13까지 1일 간격으로 변경하면서 get_weather_data 함수를 호출하고, 데이터 프레임을 하나로 합치기
import numpy as np
date_range = pd.date_range(start='2010-01-01', end='2024-05-10')
df = pd.DataFrame()
for date in date_range:
    date = date.strftime('%Y-%m-%d')
    try:
        df = pd.concat([df, get_weather_data(date)], axis=0)
        df.to_csv('./data/weather_data.csv', index=False)
    except:
        pass
    
```
