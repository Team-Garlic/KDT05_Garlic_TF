```python
import pandas as pd

df = pd.read_csv("data/weather_data.csv")
df
```

```python
# date 열이 2018-09-20 ~ 2019-06-10 일까지의 데이터를 추출하기
df["date"] = pd.to_datetime(df["date"])
df.set_index("date", inplace=True)

# 위 방식으로 데이터 프레임 만들기
df1 = df.loc["2018-09-20":"2019-06-10"].describe().unstack().to_frame().T
df2 = df.loc["2019-09-20":"2020-06-10"].describe().unstack().to_frame().T
df1 = pd.concat([df1, df2])
df3 = df.loc["2020-09-20":"2021-06-10"].describe().unstack().to_frame().T
df1 = pd.concat([df1, df3])
df4 = df.loc["2021-09-20":"2022-06-10"].describe().unstack().to_frame().T
df1 = pd.concat([df1, df4])
df5 = df.loc["2022-09-20":"2023-06-10"].describe().unstack().to_frame().T
df1 = pd.concat([df1, df5])
df6 = df.loc["2023-09-20":"2024-06-10"].describe().unstack().to_frame().T
df1 = pd.concat([df1, df6])
```

```python
# 인덱스를 2019,2020,2021,2022,2023 부여
df1.index = [2019, 2020, 2021, 2022, 2023, 2024]
df1
```

```python
df1.to_csv("weather_data_2019_2023.csv", index=False)
```

```python
df1 = pd.read_csv("weather_data_2019_2023.csv")
df3 = pd.read_csv("data/의성군_마늘생산량.csv")


# 연도가 2019 ~ 2023년에 해당하는 행만 뽑아오기
df3 = df3[df3["연도"] >= 2019]
df3
# 재배면적, 10a당 생산량, 생산량, 의성군재배면적, 의성군생산량을 df1에 열 방향으로 붙이기
df1["재배면적"] = df3["재배면적"].reset_index(drop=True)
df1["10a당 생산량"] = df3["10a당 생산량"].reset_index(drop=True)
df1["생산량"] = df3["생산량"].reset_index(drop=True)
df1["의성군재배면적"] = df3["의성군재배면적"].reset_index(drop=True)
df1["의성군생산량"] = df3["의성군생산량"].reset_index(drop=True)
df1.to_csv("data2.csv", index=False)
```
